```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('cs-training.csv',index_col = 0)
df.head()
```

```python
# Check the delinquency distribution
sns.countplot(x = 'SeriousDlqin2yrs', data = df)
print('Proportion of People Who Defaulted: {}'.format(df['SeriousDlqin2yrs'].sum() / len(df)))
```

```python
# Null Values
null.val_sum = df.isnull().sum()
```

```python
df['RevolvingUtilizationOfUnsecuredLines'].describe()
```

```python
sns.distplot(df['RevolvingUtilizationOfUnsecuredLines'])
```

```python
default_pool = []
for i in range(int(df['RevolvingUtilizationOfUnsecuredLines'].max())):
    temp = df[df['RevolvingUtilizationOfUnsecuredLines'] > i]
    default_pool.append([i, temp['SeriousDlqin2yrs'].mean()]) 
default_pool            
```

```python
dis_ruul = []
for i in range(int(df['RevolvingUtilizationOfUnsecuredLines'].max())):
    temp_ = df[df['RevolvingUtilizationOfUnsecuredLines'] > i]
    dis_ruul.append([i, temp_['RevolvingUtilizationOfUnsecuredLines'].mean()]) 
dis_ruul  
```

```python
sns.lineplot(x = [i[0] for i in default_pool], y = [i[1] for i in default_pool])
plt.title('Proportion of Defaulters As RUUL Increases')
```

```python
print('Proportion of Defaulters with Total Amount of Money Owed Not Exceeding Total Credit Limit: {}'.format(
df.loc[(df['RevolvingUtilizationOfUnsecuredLines'] >= 0) & (df['RevolvingUtilizationOfUnsecuredLines'] <= 1)]['SeriousDlqin2yrs'].mean()))
```
